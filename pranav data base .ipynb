```python

This report explores patterns in student mental health using data visualization techniques.

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style='whitegrid')
```

```python

data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female'],
    'Age': [19, 20, 21, 19, 22, 20, 21, 22, 20, 21],
    'Depression_Level': ['Low', 'Moderate', 'High', 'Moderate', 'Low', 'High', 'Moderate', 'Low', 'High', 'Low'],
    'CGPA': [7.5, 8.2, 6.9, 7.0, 8.5, 6.8, 7.2, 8.0, 6.7, 8.1],
    'Internet_Usage': ['High', 'Medium', 'Low', 'High', 'Medium', 'Low', 'High', 'Medium', 'Low', 'High'],
    'Sleep_Duration': [6, 7, 5, 6, 8, 5, 7, 6, 5, 8],
    'Study_Hours': [2, 3, 1, 2.5, 4, 1.5, 3, 2, 1, 3.5]
}
df = pd.DataFrame(data)
df.head()
```

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#ff9999'])
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Depression_Level', data=df, palette='pastel')
plt.title('Students by Depression Level')
plt.xlabel('Depression Level')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Depression_Level', y='Age', data=df, palette='coolwarm')
plt.title('Age vs Depression Level')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='Internet_Usage', data=df, palette='muted')
plt.title('Internet Usage Count')
plt.xlabel('Internet Usage')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Gender', y='CGPA', data=df, palette='Set2')
plt.title('CGPA Distribution by Gender')
plt.show()

```

```python
plt.figure(figsize=(7,5))
sns.histplot(df['Sleep_Duration'], kde=True, bins=10, color='lightgreen')
plt.title('Sleep Duration Distribution')
plt.xlabel('Hours')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Study_Hours', y='CGPA', data=df, hue='Depression_Level', palette='deep')
plt.title('Study Hours vs CGPA')
plt.show()
```

```python
###  Conclusion
- Gender was almost evenly distributed.
- Age and sleep duration showed visible impact on depression.
- Students with more study hours often had better CGPA.
- Depression levels showed patterns with academic and lifestyle factors.
```

```python
plt.figure(figsize=(10,7))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
